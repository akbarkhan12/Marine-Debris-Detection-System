```python
from ultralytics import YOLO
import cv2
import os
from pathlib import Path
```

```python
# Set dataset directory path
dataset_dir = "F:/Balaji/MARINE/"

# Example structure
train_images = os.path.join(dataset_dir, "train/images")
val_images = os.path.join(dataset_dir, "valid/images")
train_labels = os.path.join(dataset_dir, "train/labels")
val_labels = os.path.join(dataset_dir, "valid/labels")
```

```python
# Create the YAML file for the dataset
yaml_content = """




train: F:/Balaji/MARINE/train/images
val: F:/Balaji/MARINE/valid/images
test: F:/Balaji/MARINE/test/images

nc: 10
names: ['Accessories', 'Cutlery', 'Glass', 'Litter', 'Metal', 'Natual-Debris', 'Plastic', 'Rope', 'Rubber', 'Wood']


"""

# Save the YAML file
with open("data.yaml", "w") as f:
    f.write(yaml_content)
```

```python

model = YOLO('yolov10n.pt')
```

```python

model.train(data='data.yaml', epochs=350, imgsz=640, batch=26)
```

```python
# Evaluate the trained model on validation data
metrics = model.val()
print(metrics)
```

```python
#Image code


from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO(r'D:/SHIVAA/CODES/MARINE_DEBRIS/CODE/MAIN_CODE/runs/detect/train4/weights/best.pt')

# Input and output image paths
input_image_path = r"D:/SHIVAA/CODES/MARINE_DEBRIS/CODE/MAIN_CODE/test/images/1bc7-iudfmpmn7245599_jpg.rf.616a393590ae0656d4754997ab8d5f0c.jpg"
output_image_path = r"D:/SHIVAA/CODES/MARINE_DEBRIS/CODE/MAIN_CODE/output.jpg"

# Load the image
image = cv2.imread(input_image_path)

# Perform detection
results = model.predict(image)

# Annotate detections
for result in results[0].boxes.data.tolist():
    x1, y1, x2, y2, conf, cls = result
    x1, y1, x2, y2 = map(int, (x1, y1, x2, y2))  # Bounding box coordinates
    label = f"{model.names[int(cls)]} {conf:.2f}"  # Label with confidence

    # Draw bounding box and label
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box
    cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Save the output image
cv2.imwrite(output_image_path, image)


```
